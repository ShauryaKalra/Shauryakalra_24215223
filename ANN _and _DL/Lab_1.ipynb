# Lab 1: Simulating a Perceptron using NumPy

A **Perceptron** is the simplest form of a neural network — a single neuron that:
1. Takes weighted inputs
2. Sums them (+ bias)
3. Applies a step activation function
4. Outputs 0 or 1

The weight update rule (learning rule):
$$w = w + \eta \cdot (y - \hat{y}) \cdot x$$

```python
import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.1, epochs=10):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None

    def step_function(self, z):
        return 1 if z >= 0 else 0

    def predict(self, x):
        z = np.dot(x, self.weights) + self.bias
        return self.step_function(z)

    def train(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        print(f"{'Epoch':<8}{'Weights':<30}{'Bias':<10}{'Errors'}")
        print("-" * 60)

        for epoch in range(self.epochs):
            total_error = 0
            for xi, yi in zip(X, y):
                y_pred = self.predict(xi)
                error = yi - y_pred
                self.weights += self.lr * error * xi
                self.bias   += self.lr * error
                total_error += int(error != 0)
            print(f"{epoch+1:<8}{str(np.round(self.weights,3)):<30}{self.bias:<10.3f}{total_error}")
            if total_error == 0:
                print("\nConverged early!")
                break
```

## Example 1: AND Gate

```python
# AND Gate truth table
X_and = np.array([[0, 0],
                  [0, 1],
                  [1, 0],
                  [1, 1]])
y_and = np.array([0, 0, 0, 1])

p_and = Perceptron(learning_rate=0.1, epochs=10)
print("Training on AND Gate:")
p_and.train(X_and, y_and)

print("\nPredictions after training:")
for x, label in zip(X_and, y_and):
    print(f"  Input: {x}  Predicted: {p_and.predict(x)}  Actual: {label}")
```

## Example 3: NAND Gate

```python
# NAND Gate truth table
X_nand = np.array([[0, 0],
                   [0, 1],
                   [1, 0],
                   [1, 1]])
y_nand = np.array([1, 1, 1, 0])

p_nand = Perceptron(learning_rate=0.1, epochs=10)
print("Training on NAND Gate:")
p_nand.train(X_nand, y_nand)

print("\nPredictions after training:")
for x, label in zip(X_nand, y_nand):
    print(f"  Input: {x}  Predicted: {p_nand.predict(x)}  Actual: {label}")
```

## Note: XOR Gate — Not Linearly Separable
A single perceptron **cannot** learn the XOR function because XOR is not linearly separable. This limitation motivated the development of multi-layer networks (MLPs).

```python
# XOR Gate — perceptron will FAIL to converge
X_xor = np.array([[0, 0],
                  [0, 1],
                  [1, 0],
                  [1, 1]])
y_xor = np.array([0, 1, 1, 0])

p_xor = Perceptron(learning_rate=0.1, epochs=10)
print("Training on XOR Gate (expected to fail):")
p_xor.train(X_xor, y_xor)

print("\nPredictions after training:")
for x, label in zip(X_xor, y_xor):
    print(f"  Input: {x}  Predicted: {p_xor.predict(x)}  Actual: {label}")
```
