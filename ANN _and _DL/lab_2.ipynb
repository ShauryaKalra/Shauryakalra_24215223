# Lab 2 — Single Neuron Model (Manual Implementation)
Implementing a single artificial neuron from scratch using only Python and NumPy.

```python
import numpy as np

# ---------- Activation Function ----------
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# ---------- Single Neuron ----------
class Neuron:
    def __init__(self, n_inputs, lr=0.1, epochs=1000):
        self.w  = np.zeros(n_inputs)   # weights
        self.b  = 0.0                  # bias
        self.lr = lr
        self.epochs = epochs

    def predict(self, X):
        z = np.dot(X, self.w) + self.b
        return sigmoid(z)

    def train(self, X, y):
        m = len(y)
        for epoch in range(self.epochs):
            y_hat = self.predict(X)
            error = y_hat - y

            # Gradient descent update
            self.w -= self.lr * (1/m) * np.dot(X.T, error)
            self.b -= self.lr * (1/m) * np.sum(error)

            if (epoch + 1) % 200 == 0:
                loss = -np.mean(y * np.log(y_hat + 1e-8) + (1-y) * np.log(1 - y_hat + 1e-8))
                print(f"Epoch {epoch+1:4d} | Loss: {loss:.4f}")
```

```python
# ---------- AND Gate Dataset ----------
X = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=float)
y = np.array([0, 0, 0, 1], dtype=float)    # AND truth table

neuron = Neuron(n_inputs=2, lr=0.5, epochs=1000)
neuron.train(X, y)
```

```python
# ---------- Results ----------
print("\nFinal weights:", neuron.w)
print("Final bias  :", neuron.b)
print("\nPredictions (threshold 0.5):")
for xi, yi in zip(X, y):
    prob = neuron.predict(xi)
    pred = int(prob >= 0.5)
    print(f"  Input {xi.astype(int)} → prob={prob:.3f}  predicted={pred}  actual={int(yi)}")
```
